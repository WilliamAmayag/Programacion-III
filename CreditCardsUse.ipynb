```python
import pandas as guerreroDragon
import matplotlib.pyplot as meGraficas

try:
  tarjetasCredito = guerreroDragon.read_csv("/content/trx_tarjetas_credito.csv", sep=",")
  print("Archivo cargado correctamente")
except:
  print('Por Favor subir el archivo con el nombre:\n trx_tarjetas_credito.csv')

```

```python
getGender = guerreroDragon.crosstab(index=tarjetasCredito["Gender"],columns="Gender")
print(getGender)

pieGender = getGender.loc[getGender.index]["Gender"]
meGraficas.pie(pieGender, labels=pieGender.index)
meGraficas.title("Uso de tarjetas por genero")
meGraficas.show

getGender.to_excel("Credit_Card_For_Gender.xlsx")
```

```python
getCityType = tarjetasCredito.groupby(["City", "Card Type"]).sum()
print(getCityType)

pieCityType = getCityType.loc[getCityType[:10].index]
meGraficas.pie(pieCityType, labels=pieCityType.index)
meGraficas.title("Uso de tarjetas por ciudades")
meGraficas.show

getCityType.to_excel("Credit_Card_For_City.xlsx")
```

```python
getUse = guerreroDragon.crosstab(index=tarjetasCredito["Exp Type"],columns="Use")
print(getUse)

pieUse = getUse.loc[getUse.index]["Use"]
meGraficas.pie(pieUse, labels=pieUse.index)
meGraficas.title("Uso de tarjetas por gasto")
meGraficas.show

getUse.to_excel("Credit_Card_For_Use.xlsx")
```
